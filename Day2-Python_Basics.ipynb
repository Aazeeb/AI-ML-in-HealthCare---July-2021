```python
#assignment
#ask user for the season(summer, winter, rainy)
# 1. summer - it is hot
#2. winter: it is cold
#3. rainy- it is monsoon time
```

```python
season= input("Enter the season(summer, winter or Rainy): ")
season=season.lower()

if season == 'summer':    
  print("It is hot")
elif season == 'winter':     #if there are multiple ifs, every other if statements apart from the 1st 
  print("It is cold ")       #should be treated as else if, in python, else if is listed as elif
elif season == "rainy":
  print("It is monsoon time")
else:
  print("wrong season name entered")
```
