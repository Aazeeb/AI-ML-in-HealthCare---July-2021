```python
%%writefile app.py
import streamlit as st
import pandas as pd
st.title("DNA example")
# st.image("/content/dna.jpg")
st.write("""
DNA Nucleotide Count WebApp
This app counts the nucleotide compositon of query DNA!
""")
st.header('Enter DNA Sequence')
sequence=st.text_area("Sequence input", height=250)
st.write("""
****
""")
st.header('OUTPUT(DNA Nucleotide Count)')
st.write('There are ' + str(sequence.count('A')) + ' Adenine(A)')
st.write('There are ' + str(sequence.count('T')) + ' Thymine(T)')
st.write('There are ' + str(sequence.count('G')) + ' Guanine(G)')
st.write('There are ' + str(sequence.count('C')) + ' Cytosine(C)')

df = pd.DataFrame({'A':[sequence.count('A')],'T':[sequence.count('T')],'G':[sequence.count('G')],'C':[sequence.count('C')]})
st.write(df)
```
